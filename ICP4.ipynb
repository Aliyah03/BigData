```python
#1 - Pandas

#Read the provided CSV file ‘data.csv’
from google.colab import drive
drive.mount('/content/gdrive', force_remount=True)
import pandas as pd
df=pd.read_csv('gdrive/My Drive/Colab Notebooks/Big Data Analytics/data.csv')
print(df)

#Show the basic statistical description about the data
mean = df.mean()
print("\nMean Values:\n", mean)

#Check if the data has null values.
  #Replace the null values with the mean
df.fillna(mean, inplace=True)

#Select at least two columns and aggregate the data using: min, max, count, mean
dMin1 = df["Duration"].min()
dMax1 = df["Duration"].max()
dCount1 = df["Duration"].count()
dMean1 = df["Duration"].mean()
print("\nDuration Column")
print("Min: ", dMin1, "  Max: ", dMax1, "  Count: ", dCount1, "  Mean: ", dMean1)

dMin2 = df["Calories"].min()
dMax2 = df["Calories"].max()
dCount2 = df["Calories"].count()
dMean2 = df["Calories"].mean()
print("\nCalories Column")
print("Min: ", dMin2, "  Max: ", dMax2, "  Count: ", dCount2, "  Mean: ", dMean2)
print("\n")

#Filter the dataframe to select the rows with calories values between 500 and 1000
filtered = df[(df['Calories'] > 499) & (df['Calories'] < 1001)]
print("Filtered Data between 500 & 1000:")
print(filtered)
print("\n")

#Filter the dataframe to select the rows with calories values > 500 and pulse < 100
filtered2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
print("Filtered Data w/ calories > 500 & pulse < 1000:")
print(filtered2)

# Create a new “df_modified” dataframe that contains all the columns from df except for “Maxpulse”
df2 = df[['Duration', 'Pulse', 'Calories']].copy()

#Delete the “Maxpulse” column from the main df dataframe
del df["Maxpulse"]

#Convert the datatype of Calories column to int datatype
df["Calories"] = df["Calories"].astype(float).astype(int)

#Using pandas create a scatter plot for the two columns (Duration and Calories)
ax1 = df.plot.scatter(x='Duration', y='Calories')

```
